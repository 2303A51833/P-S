```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/sample_data/cricket_dataset.csv")
print(df)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df["Runs"],bins=10,kde=True,color='blue')#kernel Density # Changed 'true' to 'True'
plt.title(f'Histogram of {"Runs"}')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))


sns.scatterplot(x=df["Runs"], y=df["Matches"], color='Red')
plt.title(f'scatterplot of {"Runs"} vs {"Matches"}')
plt.xlabel('Runs')
plt.ylabel('Matches')
plt.show()
```

```python
#bar graph
plt.figure(figsize=(10,6))
sns.barplot(x=df["Player"],y=df["50s"],palette='viridis')
plt.title(f'Barplot of {"Player"}by{"50s"}')
plt.xlabel('Player')
plt.ylabel('50s')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df["Wickets"],bins=10,kde=True,color='green')#kernel Density # Changed 'true' to 'True'
plt.title(f'Histogram of {"Wickets"}')
plt.xlabel('Wickets')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))


sns.scatterplot(x=df["Wickets"], y=df["Economy Rate"], color='black')
plt.title(f'scatterplot of {"Wickets"} vs {"Economy Rate"}')
plt.xlabel('Wickets')
plt.ylabel('Economy rate')
plt.show()
```

```python
#bar graph
plt.figure(figsize=(10,6))
sns.barplot(x=df["Player"],y=df["50s"],palette='viridis')
plt.title(f'Barplot of {"Player"}by{"100s"}')
plt.xlabel('Player')
plt.ylabel('100s')
plt.xticks(rotation=45)
plt.show()
```
